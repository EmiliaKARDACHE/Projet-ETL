# Lecture et aperçu du dataset

```python
import pandas as pd
import os
```

```python
# Construction du chemin vers le CSV à partir du notebook
csv_path = os.path.join("..", "data", "processed", "agribalyse_detail_par_etape.csv")

# Lecture du CSV
df = pd.read_csv(csv_path, sep=",", encoding="utf-8-sig")

# Affichage des 5 premières lignes
df.head()
```

# 1°) Nettoyage des données

### A) Suppression des doublons

```python
# Nombre de lignes avant suppression des doublons
n_before = len(df)
print(f"Nombre de lignes avant suppression des doublons : {n_before}")

# Identifier les doublons exacts
duplicates = df[df.duplicated(keep=False)]  # keep=False marque tous les doublons

# Afficher les doublons s'il y en a
if not duplicates.empty:
    print(f"\n⚠️ {len(duplicates)} lignes dupliquées détectées :")
    print(duplicates)
else:
    print("\n✅ Aucun doublon exact détecté.")

# Suppression des doublons exacts
df = df.drop_duplicates()

# Nombre de lignes après suppression
n_after = len(df)
print(f"\nNombre de lignes après suppression des doublons : {n_after}")
print(f"Nombre de lignes supprimées : {n_before - n_after}")
```

### B) Gestion des valeurs manquantes

```python
# Dictionnaire des correspondances "mot-clé" → "unité"
unit_mapping = {
    "Score unique EF": "mPt/kg de produit",
    "Changement climatique": "kg CO2 eq/kg de produit",
    "Appauvrissement de la couche": "kg CVC11 eq/kg de produit",
    "Rayonnements ionisants": "kBq U-235 eq/kg de produit",
    "Formation photochimique": "kg NMVOC eq/kg de produit",
    "Particules fines": "disease inc./kg de produit",
    "Effets toxicologiques sur la santé humaine : substances non-cancérogènes": "CTUh/kg de produit",
    "Effets toxicologiques sur la santé humaine : substances cancérogènes": "CTUh/kg de produit",
    "Acidification terrestre et eaux douces": "mol H+ eq/kg de produit",
    "Eutrophisation eaux douces": "kg P eq/kg de produit",
    "Eutrophisation marine": "kg N eq/kg de produit",
    "Eutrophisation terrestre": "mol N eq/kg de produit",
    "Écotoxicité pour écosystèmes aquatiques": "CTUe/kg de produit",
    "Utilisation du sol": "Pt/kg de produit",
    "Épuisement des ressources eau": "m3 depriv./kg de produit",
    "Épuisement des ressources énergétiques": "MJ/kg de produit",
    "Épuisement des ressources minéraux": "kg Sb eq/kg de produit",
    "Changement climatique - émissions biogéniques": "kg CO2 eq/kg de produit",
    "Changement climatique - émissions fossiles": "kg CO2 eq/kg de produit",
    "Changement climatique - émissions liées au changement": "kg CO2 eq/kg de produit"
}

# Création d'une liste pour stocker les nouveaux noms de colonnes
new_columns = []

for col in df.columns:
    new_col = col  # par défaut, on garde le nom original
    
    for keyword, unit in unit_mapping.items():
        if keyword.lower() in col.lower():  # comparaison insensible à la casse
            new_col = f"{col} ({unit})"
            break  # on arrête dès qu’on trouve une correspondance
    
    new_columns.append(new_col)

# Appliquer les nouveaux noms de colonnes au DataFrame
df.columns = new_columns

# Vérification
print("✅ Colonnes renommées avec unités :")
print(df.columns.tolist())
```

```python
# --- Sauvegarde du DataFrame nettoyé ---
output_path = os.path.join("..", "data", "processed", "agribalyse_detail_par_etape.csv")
df.to_csv(output_path, sep=",", index=False, encoding="utf-8-sig")

print(f"✅ Nouveau fichier CSV créé : {output_path}")
```
